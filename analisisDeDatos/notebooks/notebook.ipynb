```python
import pandas as pd

import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df=pd.read_csv("../autorizaciones.csv")

valores_y=df["Cantidad Autorizaciones"]
valores_x=df["Año"]

plt1.bar(valores_x,valores_y)
plt1.title("Autorizaciones otrogadas por año")
plt1.xlabel("Año")
plt1.ylabel("Cantidad de autorizaciones")
plt1.show()

plt2.plot(valores_x,valores_y)
plt2.title("Autorizaciones otrogadas por año")
plt2.xlabel("Año")
plt2.ylabel("Cantidad de autorizaciones")
plt2.show()
```

# Conclusión
#### La cantidad de autorizaciones otorgadas aumentaron 8 veces su valor en un período de 5 años
