```python
import pandas as pd

import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

casos=pd.read_csv("../casos.csv",sep=",",engine="python")

print(len(casos))

casos=casos.dropna(subset=['barrio'])

valores_x=casos.barrio.drop_duplicates().tolist()
print(len(valores_x))
valores_y=casos.barrio.value_counts().tolist()

plt1.bar(valores_x,valores_y)
plt1.title("casos por barrio")
plt1.xlabel("barrio")
plt1.ylabel("Cantidad de casos")
plt1.show()

plt2.plot(valores_x,valores_y)
plt2.title("Autorizaciones otrogadas por año")
plt2.xlabel("Año")
plt2.ylabel("Cantidad de autorizaciones")
plt2.show()
```

# Conclusión
#### La cantidad de autorizaciones otorgadas aumentaron 8 veces su valor en un período de 5 años
