```python
import pandas as pd

import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

casos=pd.read_csv("../casos.csv",sep=",",engine="python")

casos=casos.dropna(subset=['barrio'])


b=casos["barrio"].value_counts().rename_axis('unique_values').reset_index(name='counts')
plt2.plot(b["unique_values"],b["counts"])
plt2.title("Cantidad de casos por barrio")
plt2.xlabel("BARRIOS")
plt2.xticks(rotation=90)
plt2.show()
#casos.barrio.value_counts().sort_values().plot(kind = 'bar')
```

```python
poblacionRiesgo=casos[(casos.barrio=="PALERMO") | (casos.barrio=="BELGRANO")]
ax=poblacionRiesgo["edad"].unique()
ay=poblacionRiesgo["edad"].value_counts().tolist()
ay.reverse()
plt1.bar(ax,ay)
plt1.xlabel("EDAD")
plt1.show()

```

# Conclusin
#### La mayor cantidad de casos se encuentran en los barrios Palermo y Belgrano. Se analizaron las edades de quienes dieron positivo en estos sectores de la ciudad y se concluy que la mayora no se encuentran en el rango de edad considerado poblacin de riesgo. De todas formas, no se puede analizar si la poblacin mostrada en estos grficos tiene alguna enfermedad previa ya que esos datos no figuran en el dataset, por lo que no se puede estar 100% seguros de que no pertenecen a ningn grupo de riesgo.
