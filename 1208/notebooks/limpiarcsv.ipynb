```python
import pandas as pd

import numpy as np

datos=pd.read_csv("datosALimpiar.csv")

def arreglarEspacios(valor):
    valor = str(valor)
    if valor[0] == ' ':
        valor = valor[1:]
    if valor == '':
        valor = np.nan
    if valor == 'nan':
        valor = np.nan
    return valor

def arreglarObservacion(valor):
    if "hiper" not in valor and "cabeza" not in valor:
        valor= "ninguna"
    elif "cabeza" in valor:
        valor= "dolor de cabeza"
    else:
        valor="hipertensi贸n"
    return valor

def arreglarLu(valor):
    valor = str(valor)
    if '//' in valor:
        valor = valor.replace('//','/')
    for caracter in valor:
        if not caracter.isnumeric() and caracter != "/":
          valor = valor.replace(caracter,'/')

    return valor

def arreglarSexo(valor):
    valor = str(valor)
    if valor == "No binarie":
        valor="nb"
    elif "H" in valor:
        valor="h"
    elif "M" in valor:
        valor="m"
    return valor
def arreglarPeso(valor):
    valor=str(valor)
    valor=valor.replace(' kg','')
    return valor

def arreglarAltura(valor):
    valor=valor.replace('m','')
    return valor

datos=datos.applymap(arreglarEspacios)
datos=datos.dropna()
print(datos)
datos[' Observacion']=datos[' Observacion'].apply(arreglarObservacion)
datos[' LU']=datos[' LU'].apply(arreglarLu)
datos=datos.loc[(datos[" Sexo"].str.contains("M")) | (datos[" Sexo"].str.contains("H")) | (datos[" Sexo"].str.contains("No"))]
datos[' Sexo']=datos[' Sexo'].apply(arreglarSexo)
datos[' Peso']=datos[' Peso'].apply(arreglarPeso)
datos[' Altura']=datos[' Altura'].apply(arreglarAltura)
print(datos)
```
