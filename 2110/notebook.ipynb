```python
import pandas as pd
import seaborn as sns


import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import numpy as np


df=pd.read_csv("historico_temperaturas.csv")

max=df.iloc[df['máxima'].idxmax()]
min=df.iloc[df['mínima'].idxmin()]

print("La temperatura máxima que fué registrada")
print(max)
print("La temperatura mínima que fué registrada")
print(min)

def mesMinMax(anio):
    print('\n')
    print("Se calcula el mes más caluroso y el más frio del año")
    print(anio)
    soloAnio=df.loc[df['año'] == anio]
    max=soloAnio.iloc[soloAnio['media'].idxmax()]
    min=soloAnio.iloc[soloAnio['media'].idxmin()]

    mesMin=max.loc["mes"]
    mesMax=min.loc["mes"]
    tempMax=max.loc["media"]
    tempMin=min.loc["media"]
    print("Mes más caluroso",mesMax)
    print("Con la media de",tempMax)
    print("Mes más frío",mesMin)
    print("Con la media de",tempMin)

mesMinMax(1991)


df2 = df[["máxima","mínima","año","mes"]]
sns.pairplot(df2,hue="mes",palette="Spectral")
```

### Con estos gráficos se puede apreciar la tendencia de las temperaturas de cada mes a aumentar pero no considerablemente.
